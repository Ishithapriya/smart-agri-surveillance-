RESIZE

```python
import cv2

img = cv2.imread("myimage.png")

resized = cv2.resize(img, (500, 300))

cv2.imshow("Resized Image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

GRAYSCALE

```python
import cv2

img = cv2.imread("myimage.png")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Grayscale Image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

OPEN WEBCAM

```python
import cv2

cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Cannot read camera")
        break

    cv2.imshow("Camera", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

DATE AND TIME

```python
import cv2
from datetime import datetime
cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    cv2.putText(
        frame,
        now,
        (10, 30),  # position (x, y)
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,        # font scale
        (0, 255, 0), # color (green)
        2,          # thickness
        cv2.LINE_AA
    )

    cv2.imshow("Webcam with DateTime", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

DRAW RECTANGLE

```python
import cv2
cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.rectangle(frame, (100, 100), (300, 200), (255, 0, 0), 2)
    cv2.putText(
        frame,
        "Live Camera",
        (50, 40),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2,
        cv2.LINE_AA
    )
    cv2.imshow("Webcam", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

SAVING

```python
import cv2
cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to access camera")
        break

    cv2.imshow("Press 's' to Save, 'q' to Quit", frame)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('s'):
        filename = "captured_image.jpg"
        cv2.imwrite(filename, frame)
        print(f"Image saved as {filename}")

    if key == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
